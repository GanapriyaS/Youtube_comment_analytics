```python
def diagnosis(row):
        if(row['text'].lower().__contains__('nice') or row['text'].lower().__contains__('thank') or row['text'].lower().__contains__('good') or row['text'].lower().__contains__('tqq') or row['text'].lower().__contains__('tqs')  or row['text'].lower().__contains__('super') or row['text'].lower().__contains__('good') or row['text'].lower().__contains__('tqq') or row['text'].lower().__contains__('tqs')  or row['text'].lower().__contains__('excellent') or row['text'].lower().__contains__('good') or row['text'].lower().__contains__('tqq') or row['text'].lower().__contains__('tqs')  or row['text'].lower().__contains__('amazing') or row['text'].lower().__contains__('good') or row['text'].lower().__contains__('tqq') or row['text'].lower().__contains__('tqs')  or row['text'].lower().__contains__('great') or row['text'].lower().__contains__('good') or row['text'].lower().__contains__('tqq') or row['text'].lower().__contains__('tqs')  or row['text'].lower().__contains__('heaven')   ):
             return 0
            
        elif(row['text'].lower().__contains__('plz') or row['text'].lower().__contains__('please') or row['text'].lower().__contains__('pls') ):
             return 2
        elif(row['text'].lower().__contains__('bad') or row['text'].lower().__contains__('worst') or row['text'].lower().__contains__('waste') ):
             return 4
        else:
            return row['value']

df['value'] = df.apply(diagnosis, axis=1)
```
